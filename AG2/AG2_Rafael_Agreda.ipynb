```
#Decorador para calcular_tiempo a la función para la cual queremos calcular el tiempo de ejecución(Isabel Vázquez)

from time import time
#Función para calcular el tiempo de ejecución
def calcular_tiempo(f):
   
    def wrapper(*args, **kwargs):        
        inicio = time()       
        resultado = f(*args, **kwargs)       
        tiempo = time() - inicio
        print("Tiempo de ejecución para algoritmo: "+str(tiempo))
        return resultado
    
    return wrapper
from time import time
```

```
import random
import math

N=4000

Lista_2D = [ ((random.randrange(1,N*10)),(random.randrange(1,N*10))) for _ in range(N)]
print(Lista_2D)
```

```
def distancia(A,B):
  if type(A) is int or type(A) is float:
    return abs(B-A)
  else:
    return math.sqrt(sum([(A[i]-B[i])**2 for i in range(len(A))]))
 

distancia((1,3),(2,6))
```

```
#Fuerza Bruta
#@calcular_tiempo
def distancia_fuerza_bruta(L):
  mejor_distancia=100000e10
  A,B=(),()
  for i in range(len(L)):
    for j in range(i+1, len(L)):
       D=distancia(L[i],L[j]) 
       if D < mejor_distancia:
                A,B=L[i],L[j]
                mejor_distancia = D
  return [A,B] 

distancia_fuerza_bruta(Lista_2D) 
```

```
#Fuerza Bruta 1 La hice aplicando el for asignado directamente a la lista (Cosa que estaba acostumbrado hacer porque se veía mejor para mi)
#El tiempo de ejecución fue el doble, imagino hace una asignación en cada iteración (17 SEG VS LOS 34 DE ESTE)
@calcular_tiempo
def distancia_fuerza_bruta1(L):
  mejor_distancia=100000e10
  A,B=(),()
  for i in L:
    for j in L:
       D=distancia(i,j) 
       if i!=j and D < mejor_distancia:
                A,B=i,j
                mejor_distancia = D
  return [A,B] 

distancia_fuerza_bruta1(Lista_2D) 
```

```
#Divide y venceras
def distancia_divide_y_venceras(L):
  #si hay pocos por fuerza bruta
  if len(L)<10:
    return distancia_fuerza_bruta(L)
  
  #Dividir en listas grandes
  #pivote = sum([L[i][0] for i in range(len(L))])/len(L)
  #print(pivote)
  
  #APORTE
  #ORDENO PRIMERO LA LISTA EN X
  XL=sorted(L, key=lambda x: x[0])
    
  #DIVIDO EN X y en Y
  X_LISTA_IZQ = XL[:len(XL)//2]
  X_LISTA_DER = XL[len(XL)//2:]
  
  XPUNTOS_LISTA_IZQ = distancia_divide_y_venceras(X_LISTA_IZQ)
  XPUNTOS_LISTA_DER = distancia_divide_y_venceras(X_LISTA_DER)  
  
  #Calcula la primera distancia
  X_distancia = distancia_fuerza_bruta(XPUNTOS_LISTA_IZQ + XPUNTOS_LISTA_DER)
  
  #Ordeno la lista en Y y divido
  YL=sorted(L, key=lambda x: x[1])
  Y_LISTA_IZQ = YL[:len(YL)//2]
  Y_LISTA_DER = YL[len(YL)//2:]

  YPUNTOS_LISTA_IZQ = distancia_divide_y_venceras(Y_LISTA_IZQ)
  YPUNTOS_LISTA_DER = distancia_divide_y_venceras(Y_LISTA_DER)

  Y_distancia = distancia_fuerza_bruta(YPUNTOS_LISTA_IZQ + YPUNTOS_LISTA_DER)
  
  #Tomando en cuenta la distancia de las dos divisiones tomo la menor
  if distancia(X_distancia[0],X_distancia[1]) > distancia(Y_distancia[0],Y_distancia[1]):
    return Y_distancia
  else:
    return X_distancia
  
@calcular_tiempo  
def LANZA(L):
  return distancia_divide_y_venceras(L)
  
SOL = LANZA(Lista_2D[:1000])

print(SOL)
  
#distancia_divide_y_venceras(Lista_2D[:20])
```

```
#paseo por el rio
def precios(TARIFAS):
  N = len(TARIFAS)
  PRECIOS = [[999]*N for i in [9999]*N]
  RUTAS = [[""]*N for i in [9999]*N]
  
 
  for i in range(N-1):
    for j in range(i+1,N):
      MIN = TARIFAS[i][j]
      RUTAS[i][j] = i
      
      for k in range(i,j):
        if PRECIOS[i][k]+ TARIFAS[k][j] < MIN:
          MIN = min( MIN , PRECIOS[i][k]+ TARIFAS[k][j] )
          RUTAS[i][j] = k
      PRECIOS[i][j] = MIN
      
  return PRECIOS, RUTAS

  
PRECIOS, RUTAS = precios(TARIFAS)

print(PRECIOS)

print()

print(RUTAS)

def calcular_ruta(RUTAS, desde, hasta):
  if desde == hasta:
    #print("Ir a :" + str(desde))
    return desde 
  else:
    return str(calcular_ruta(RUTAS, desde, RUTAS[desde][hasta])) +  ',' + str(RUTAS[desde][hasta]) 

print("\nLa ruta es:")  
calcular_ruta(RUTAS, 0,6)

precios(TARIFAS)
```
